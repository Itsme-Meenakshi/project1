# Detecting Dogs and Cats



```python
import cv2
import numpy as np
import tensorflow
from tensorflow.keras.preprocessing import image
import matplotlib.pyplot as plt
```

```python
classifier = tensorflow.keras.models.load_model('dogcat_model_bak.h5')

def preprocess_frame(frame, target_size=(64, 64)):
    img = cv2.resize(frame, target_size)
    img = img.astype('float32') / 255.0
    img = np.expand_dims(img, axis=0)  # create a batch of size 1 [N,H,W,C]
    return img

cap = cv2.VideoCapture(0)  

if not cap.isOpened():
    print("Error: Could not open webcam.")
    exit()
```

```python
while True:
    ret, frame = cap.read()
    if not ret:
        print("Failed to grab frame.")
        break
    
    img = preprocess_frame(frame)
    
    prediction = classifier.predict(img) 
    
    if prediction[0, 0] > 0.5:
        label = f"Dog: {prediction[0, 0]:.2f}"
        color = (0, 0, 255)  # Red for Dog
        
    else:
        label = f"Cat: {1.0 - prediction[0, 0]:.2f}"
        color = (255, 0, 0)  # Blue for Cat
    
    
    cv2.putText(frame, label, (20, 50), cv2.FONT_HERSHEY_SIMPLEX, 1, color, 2, cv2.LINE_AA)
    
    # Show the frame
    cv2.imshow('Real-Time Prediction', frame)
    
    # Break loop on 'q' key press
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

cap.release()
cv2.destroyAllWindows()
```
